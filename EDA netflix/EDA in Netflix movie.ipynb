## highest number of movies and Tvshows by country

```python

movie = Netflix[Netflix['type']=='Movie']

counts = movie['country'].value_counts().head(10)

low_counts = movie['country'].value_counts().tail(10)

fig, ax = plt.subplots(figsize=(12, 6))

# Bar plot for the top 10 countries with the highest number of movies released
counts.plot(kind='bar', color='purple', ax=ax)
ax.set_title('highest number of movies released in a country ')
ax.set_xlabel('Country')
ax.set_ylabel('Count')

fig, ax = plt.subplots(figsize=(12, 6))

# Bar plot for the top 10 countries with the highest number of movies released
low_counts.plot(kind='bar', color='blue', ax=ax)
ax.set_title('Lowest amount of movies released in a country')
ax.set_xlabel('Country')
ax.set_ylabel('Count')
```

```python
## for tv shows
Tv = Netflix[Netflix['type']=='TV Show']

counts_t = Tv['country'].value_counts().head(10)

low_counts_t = Tv['country'].value_counts().tail(10)

fig, ax = plt.subplots(figsize=(12, 6))

# Bar plot for the top 10 countries with the highest number of movies released
counts_t.plot(kind='bar', color='red', ax=ax)
ax.set_title('highest number of Tvshows released in a country ')
ax.set_xlabel('Country')
ax.set_ylabel('Count')

fig, ax = plt.subplots(figsize=(12, 6))

# Bar plot for the top 10 countries with the highest number of movies released
low_counts_t.plot(kind='bar', color='yellow', ax=ax)
ax.set_title('Lowest amount of Tv shos released in a country')
ax.set_xlabel('Country')
ax.set_ylabel('Count')
```

```python
total_count = Netflix['release_year'].value_counts()

Highest_year = total_count.head(10)
lowest_year = total_count.tail(10)

plt.figure(figsize =(14, 6))

plt.subplot(1, 2, 1)
plt.bar(Highest_year.index, Highest_year.values, color = 'green')
plt.xlabel('release_year')
plt.ylabel('Count')
plt.title('Combined highest number of release years for both movies and Tv shows ')

plt.subplot(1, 2, 2)
plt.bar(lowest_year.index, lowest_year.values, color = 'orange')
plt.xlabel('release_year')
plt.ylabel('Count')
plt.title('Combined lowest number of release years for both movies and Tv shows ')

plt.tight_layout()
plt.show()
```
